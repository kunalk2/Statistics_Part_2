***Q1. What is hypothesis testing in statistics ?***

Ans. Hypothesis testing in statistics is a method used to evaluate a claim about a population parameter based on sample data. It involves formulating a null hypothesis (a statement of no effect or no difference) and an alternative hypothesis (a statement that contradicts the null), collecting data, and then using statistical analysis to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative. Essentially, it helps researchers decide if observed results are likely due to chance or if they provide evidence for a real effect.

***Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?***

Ans.
1. Null Hypothesis (H₀):
* It's a statement that assumes there's no significant relationship or difference between variables being studied.
* It's often the default position that researchers try to disprove.
* It typically uses an equality symbol (e.g., =, ≥, ≤) in its statement.
* Example: "There is no difference in test scores between students who study with flashcards and those who don't."
2. Alternative Hypothesis (H₁ or Ha):
* It's a statement that directly contradicts the null hypothesis, suggesting there is a significant relationship or difference between the variables.
* It's what researchers are typically trying to prove or find evidence for.
* It typically uses inequality symbols (e.g., ≠, <, >).
* Example: "Students who study with flashcards will have significantly higher test scores than those who don't."

***Q3. What is the significance level in hypothesis testing, and why is it important ?***

Ans. In hypothesis testing, the significance level (often denoted as α or alpha) is the probability of rejecting the null hypothesis when it is actually true. It represents the threshold at which a result is deemed statistically significant and is pre-determined by the researcher before conducting the test. Essentially, it's the risk you're willing to take of incorrectly rejecting a true null hypothesis.

It is importanat as :
1. Setting the standard for evidence:
The significance level establishes the level of evidence needed to reject the null hypothesis. A lower significance level (e.g., 0.01) requires stronger evidence than a higher one (e.g., 0.05).
2. Controlling Type I errors:
A Type I error occurs when you reject the null hypothesis when it's actually true. The significance level directly controls the probability of making this type of error.
3. Balancing error risks:
Choosing the significance level involves a trade-off. A lower significance level reduces the risk of Type I errors but increases the risk of failing to reject a false null hypothesis (Type II error). A higher significance level does the opposite.
4. Guiding decision-making:
The significance level is a crucial factor in determining whether to accept or reject the null hypothesis. It helps researchers interpret their results and draw meaningful conclusions.
5. Impacts research conclusions:
The chosen significance level influences the conclusions drawn from a study. For example, in medical research, a very low significance level (e.g., 0.01) might be preferred to minimize the risk of false positives.

***Q4. What does a P-value represent in hypothesis testing  ?***

Ans. In hypothesis testing, a p-value represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. Essentially, it quantifies the evidence against the null hypothesis; a small p-value suggests strong evidence against the null, while a large p-value suggests weak evidence.

***Q5.  How do you interpret the P-value in hypothesis testing ?***

In hypothesis testing, the p-value represents the probability of observing a test statistic as extreme as, or more extreme than, the one calculated from your sample data, assuming the null hypothesis is true. Essentially, it quantifies the evidence against the null hypothesis: a small p-value suggests the observed results are unlikely under the null, leading to its rejection, while a large p-value suggests the results are likely under the null, leading to its failure to be rejected.

***Q6. What are Type 1 and Type 2 errors in hypothesis testing ?***

Ans. In hypothesis testing, Type I and Type II errors are the two types of mistakes that can occur when drawing conclusions about a population based on sample data. A Type I error (false positive) occurs when the null hypothesis is rejected despite it being true, while a Type II error (false negative) occurs when the null hypothesis is not rejected despite it being false.

It's used to estimate an unknown parameter, like a population mean, with a certain level of confidence, typically 95% or 99%. This range provides a more informative picture than a single point estimate by quantifying the uncertainty associated with the estimate.

How it's used:

* Quantifying Uncertainty:
CIs provide a measure of uncertainty associated with the point estimate. A wider CI indicates greater uncertainty and a less precise estimate, while a narrower CI suggests more precision and certainty.
* Hypothesis Testing:
CIs can be used as an alternative to p-values for hypothesis testing. If a CI for the difference between two groups (or a difference from zero) does not include zero, it suggests a statistically significant difference (at the chosen confidence level).

Here's how the margin of error affects the confidence interval:

* Margin of Error and Confidence Interval:
The margin of error is added to and subtracted from the sample estimate to create the upper and lower bounds of the confidence interval.
* Relationship:
The margin of error is directly related to the width of the confidence interval. A larger margin of error leads to a wider confidence interval, and a smaller margin of error leads to a narrower interval.
* Confidence Level:
Increasing the desired confidence level (e.g., from 95% to 99%) will generally lead to a wider confidence interval, as you need to account for a larger range of possibilities to be more confident.
* Sample Size:
A larger sample size will generally result in a smaller margin of error and, therefore, a narrower confidence interval.
* Variability in the Sample:
Greater variability in the sample (as measured by the standard deviation) will tend to increase the margin of error and widen the confidence interval.

***Q18. What is the Chi-square goodness of fit test, and how is it applied ?***

Ans. The chi-square goodness-of-fit test is a statistical test used to determine how well a sample's observed distribution matches a theoretical or expected distribution. It checks if the differences between observed and expected frequencies are due to random chance or if the sample data significantly deviates from the expected pattern.
How it's applied:
1. Define the hypothesis:
The null hypothesis (H0) assumes that the observed data follows the theoretical distribution.
2. Calculate expected frequencies:
Based on the theoretical distribution, determine the expected frequencies for each category or interval of your data.
3. Calculate the chi-square test statistic:
This statistic measures the discrepancy between observed and expected frequencies. It's calculated as: χ² = Σ [(Observed - Expected)² / Expected].
4. Determine degrees of freedom:
The degrees of freedom (df) for the chi-square test is calculated as (number of categories - 1).
5. Find the p-value:
Using the chi-square test statistic and degrees of freedom, determine the p-value, which represents the probability of obtaining the observed results (or more extreme results) if the null hypothesis is true.
6. Make a decision:
If the p-value is less than a predetermined significance level (usually 0.05), reject the null hypothesis and conclude that the observed data does not fit the theoretical distribution. If the p-value is greater than the significance level, fail to reject the null hypothesis.

***Q20. What is an ANOVA test, and what are its assumptions ?***

Ans. An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there's a statistically significant difference between them. It essentially tests the null hypothesis that all group means are equal against the alternative hypothesis that at least one group mean is different. ANOVA relies on several key assumptions about the data for the results to be valid.
Assumptions of ANOVA:
1. Independence:
Observations within each group and between groups must be independent. This means the data points should not be related or influenced by each other.
2. Normality:
The data within each group should be approximately normally distributed. This means the data should follow a bell-shaped curve, and there should be no extreme skewness or outliers.
3. Homogeneity of Variance (also known as homoscedasticity):
The variances (or standard deviations) of the groups being compared should be roughly equal. In other words, the spread of data within each group should be similar.

***Q21. What are the different types of ANOVA tests ?***

Ans. ANOVA (Analysis of Variance) tests are used to compare the means of two or more groups to determine if there's a statistically significant difference between them. There are several types of ANOVA, including one-way, two-way, and factorial ANOVA, each with its own purpose and assumptions.
1. One-Way ANOVA:
* This type of ANOVA tests for differences between the means of two or more independent groups, using only one independent variable (factor).
* It's the most basic form of ANOVA and is used to determine if the factor has a significant impact on the dependent variable.
* For example, a one-way ANOVA could be used to compare the average test scores of students using different teaching methods.

* This ANOVA examines the effects of two or more independent variables (factors) on a dependent variable.

3. Factorial ANOVA:
* Factorial ANOVA is an extension of one-way ANOVA, allowing for the analysis of multiple independent variables (factors) and their interactions.
* It's used to determine if there are main effects (individual effects of each factor) and interaction effects (combined effects of two or more factors).
* For instance, a factorial ANOVA could be used to investigate the effects of different marketing strategies and different pricing models on sales revenue.

***Q22. What is the F-test, and how does it relate to hypothesis testing ?***

Ans. The F-test is a statistical test that compares variances between two or more groups or samples. It's used in hypothesis testing to determine if there's a significant difference in the means of these groups. The F-test, often used in ANOVA (analysis of variance), calculates an F-statistic (a ratio of variances) and compares it to a critical value from the F-distribution.

F-test relates to hypothesis testing as:
1. Null and Alternative Hypotheses:
* Null hypothesis: States that there is no significant difference in the means between the groups (i.e., the variances are equal).
* Alternative hypothesis: States that there is a significant difference in the means (i.e., the variances are not equal).
2. Calculating the F-statistic:
* The F-test calculates the F-statistic, which is a ratio of the variance between groups to the variance within groups.
3. Using the F-distribution for Comparison:
* The F-statistic is then compared to the F-distribution based on the degrees of freedom (related to the number of groups and observations) and the chosen significance level (alpha).
4. Decision:
* If the calculated F-statistic is greater than the critical value from the F-distribution, the null hypothesis is rejected, suggesting that there is a significant difference in the means.
5. Interpreting the Results:
* If the null hypothesis is rejected, it indicates that the means of the groups are likely different, and further analysis (e.g., post-hoc tests) might be needed to pinpoint which specific groups differ.


```python
#Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

# Sample data
sample = [88, 92, 85, 90, 94, 91, 89, 87, 93, 95]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
sample_size = len(sample)

# Population mean (known)
population_mean = 90

# Z-test calculation
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # two-tailed test

# Output results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", sample_size)
print("Population Mean:", population_mean)
print("Z-Score:", z_score)
print("P-Value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

```python
#Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample data
sample = [102, 98, 105, 100, 103, 99, 97, 106, 101, 104]
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
sample_size = len(sample)

# Known population mean
population_mean = 100

# Z-test calculation
z_score = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Critical z value for two-tailed test at alpha = 0.05
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)

# Output results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", sample_size)
print("Population Mean:", population_mean)
print("Z-Score:", z_score)
print("Critical Z-Value:", z_critical)
print("P-Value:", p_value)

# Interpretation
if abs(z_score) > z_critical:
    print("Result: Reject the null hypothesis. Sample mean is significantly different.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference.")

# Visualization
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label="Standard Normal Distribution")
plt.axvline(z_critical, color='red', linestyle='--', label="Critical Z-Value (±{:.2f})".format(z_critical))
plt.axvline(-z_critical, color='red', linestyle='--')
plt.axvline(z_score, color='blue', linestyle='-', label="Z-Score = {:.2f}".format(z_score))
plt.fill_between(x, 0, y, where=(x > z_critical) | (x < -z_critical), color='red', alpha=0.3, label="Rejection Region")
plt.title("Two-Tailed Z-Test Decision Regions")
plt.xlabel("Z")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualizes Type I and Type II errors in hypothesis testing.

    Parameters:
    mu0: Null hypothesis mean
    mu1: Alternative hypothesis mean
    sigma: Standard deviation
    n: Sample size
    alpha: Significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value under H0
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se

    # Type I Error (alpha): Reject H0 when it's actually true
    # Type II Error (beta): Fail to reject H0 when H1 is true

    # Beta calculation (P(X < x_critical | mu = mu1))
    z_beta = (x_critical - mu1) / se
    beta = norm.cdf(z_beta)

    # Create x values for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label="H0: μ = {}".format(mu0), color='blue')
    plt.plot(x, y1, label="H1: μ = {}".format(mu1), color='green')

    # Shade Type I error region (right tail of H0)
    x_fill_alpha = x[x > x_critical]
    plt.fill_between(x_fill_alpha, 0, norm.pdf(x_fill_alpha, mu0, se), color='red', alpha=0.5, label="Type I Error (α)")

    # Shade Type II error region (left of x_critical under H1)
    x_fill_beta = x[x < x_critical]
    plt.fill_between(x_fill_beta, 0, norm.pdf(x_fill_beta, mu1, se), color='orange', alpha=0.5, label="Type II Error (β)")

    plt.axvline(x_critical, color='black', linestyle='--', label="Critical Value (x = {:.2f})".format(x_critical))
    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Critical Value (x): {x_critical:.2f}")
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.2f}")
    print(f"Power of the test (1 - β): {1 - beta:.2f}")

# Call the function with example values
visualize_type1_type2_errors()
```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default is 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound): tuple of confidence interval
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    n = len(data)
    se = sample_std / np.sqrt(n)

    # t-critical value for two-tailed test
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)

    margin_of_error = t_critical * se
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Sample Standard Deviation: {sample_std:.2f}")
    print(f"Sample Size: {n}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print("\nInterpretation:")
    print(f"We are {int(confidence*100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")

    return (lower_bound, upper_bound)

# Example usage
sample_data = [52, 55, 60, 48, 50, 54, 56, 49, 53, 51]
confidence_interval(sample_data)
```

```python
#Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayesian_inference(prior_A, prob_B_given_A, prob_B_given_not_A):
    """
    Apply Bayes' Theorem to compute the posterior probability of event A given event B.

    Parameters:
    - prior_A: P(A), the prior probability of hypothesis A
    - prob_B_given_A: P(B|A), the likelihood of observing B if A is true
    - prob_B_given_not_A: P(B|¬A), the likelihood of B if A is false

    Returns:
    - posterior: P(A|B), the updated (posterior) probability of A after observing B
    """
    prior_not_A = 1 - prior_A
    prob_B = prob_B_given_A * prior_A + prob_B_given_not_A * prior_not_A  # total probability of B

    posterior_A_given_B = (prob_B_given_A * prior_A) / prob_B

    print(f"Prior Probability P(A): {prior_A:.2f}")
    print(f"Likelihood P(B|A): {prob_B_given_A:.2f}")
    print(f"Likelihood P(B|¬A): {prob_B_given_not_A:.2f}")
    print(f"Total Probability of Evidence P(B): {prob_B:.2f}")
    print(f"Posterior Probability P(A|B): {posterior_A_given_B:.4f}")

    return posterior_A_given_B

# Example use case: Disease diagnosis
# A = person has disease
# B = test is positive

# Prior: 1% of people have the disease
# Test is 99% accurate if person has disease (True Positive)
# Test gives 5% false positives (when person doesn’t have disease)

posterior = bayesian_inference(prior_A=0.01, prob_B_given_A=0.99, prob_B_given_not_A=0.05)
```

```python
#Q16. Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

# Generate two random samples (normally distributed)
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=5, size=30)   # Sample 1 with variance ≈ 25
sample2 = np.random.normal(loc=52, scale=8, size=30)   # Sample 2 with variance ≈ 64

# Calculate sample variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# F-statistic (larger variance / smaller variance)
if var1 > var2:
    f_stat = var1 / var2
    dfn = len(sample1) - 1  # degrees of freedom numerator
    dfd = len(sample2) - 1  # degrees of freedom denominator
else:
    f_stat = var2 / var1
    dfn = len(sample2) - 1
    dfd = len(sample1) - 1

# Compute p-value for right-tailed test
p_value = 1 - f.cdf(f_stat, dfn, dfd)

# Output results
print(f"Sample 1 Variance: {var1:.2f}")
print(f"Sample 2 Variance: {var2:.2f}")
print(f"F-Statistic: {f_stat:.4f}")
print(f"Degrees of Freedom: ({dfn}, {dfd})")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The variances are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference in variances.")
```

```python
#Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

def check_anova_assumptions(*groups):
    """
    Checks the assumptions for performing ANOVA:
    1. Normality (Shapiro-Wilk Test)
    2. Independence (by plotting residuals)
    3. Homogeneity of variances (Levene’s Test)

    Parameters:
    - groups: Variable number of group arrays (e.g., group1, group2, ...)
    """
    all_data = np.concatenate(groups)
    group_labels = np.concatenate([[f'Group {i+1}'] * len(group) for i, group in enumerate(groups)])

    print("=== Normality Test (Shapiro-Wilk) for each group ===")
    for i, group in enumerate(groups):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1}: W = {stat:.4f}, p-value = {p:.4f} → {'Normal' if p > 0.05 else 'Not Normal'}")

    print("\n=== Homogeneity of Variance Test (Levene's) ===")
    levene_stat, levene_p = stats.levene(*groups)
    print(f"Levene’s Statistic = {levene_stat:.4f}, p-value = {levene_p:.4f} → {'Equal variances' if levene_p > 0.05 else 'Unequal variances'}")

    print("\n=== Independence Check: Residual Plot ===")
    # One-way ANOVA residuals (fit grand mean)
    grand_mean = np.mean(all_data)
    residuals = all_data - grand_mean

    plt.figure(figsize=(8, 4))
    plt.plot(residuals, marker='o', linestyle='-', color='blue')
    plt.axhline(0, color='black', linestyle='--')
    plt.title('Residual Plot (Assumption: Independence)')
    plt.xlabel('Observation Index')
    plt.ylabel('Residual (Y - Grand Mean)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Example usage
group1 = [88, 92, 85, 91, 87]
group2 = [78, 75, 80, 79, 77]
group3 = [90, 93, 89, 92, 91]
check_anova_assumptions(group1, group2, group3)
```

```python
#Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset: Effect of Teaching Method and Gender on Test Scores
data = {
    'Score':    [85, 88, 90, 86, 87, 83, 89, 84, 88, 85, 75, 78, 80, 76, 77, 73, 79, 74, 78, 75,
                 92, 94, 91, 90, 93, 89, 95, 92, 91, 90],
    'Method':   ['A']*10 + ['B']*10 + ['C']*10,
    'Gender':   ['Male', 'Female']*5 + ['Male', 'Female']*5 + ['Male', 'Female']*5
}

df = pd.DataFrame(data)

# Two-way ANOVA with interaction
model = ols('Score ~ C(Method) + C(Gender) + C(Method):C(Gender)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Output the ANOVA table
print("Two-Way ANOVA Table:")
print(anova_table)

# Interpretation
print("\nInterpretation:")
for factor in anova_table.index:
    p_val = anova_table.loc[factor, "PR(>F)"]
    if p_val < 0.05:
        print(f"→ '{factor}' has a significant effect (p = {p_val:.4f})")
    else:
        print(f"→ '{factor}' has no significant effect (p = {p_val:.4f})")

# Visualization: Interaction plot
plt.figure(figsize=(8, 6))
sns.pointplot(data=df, x='Method', y='Score', hue='Gender', capsize=.1, errwidth=1.5)
plt.title("Interaction Plot: Method vs Gender on Scores")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q22. Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# Sample data: Three groups (e.g., exam scores under three teaching methods)
group_A = [85, 88, 90, 86, 87]
group_B = [75, 78, 80, 76, 77]
group_C = [92, 94, 91, 90, 93]

# Combine into a single DataFrame for easier plotting
df = pd.DataFrame({
    'Score': group_A + group_B + group_C,
    'Method': ['A']*len(group_A) + ['B']*len(group_B) + ['C']*len(group_C)
})

# Perform one-way ANOVA
f_stat, p_value = stats.f_oneway(group_A, group_B, group_C)

# Display ANOVA results
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. At least one group mean is significantly different.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference between group means.")

# Visualization with boxplots
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Method', y='Score', palette="Set2")
plt.title('One-Way ANOVA: Comparison of Group Means')
plt.xlabel('Teaching Method')
plt.ylabel('Test Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(1)

# Simulate random data from a normal distribution
sample_size = 50
true_mean = 100     # population mean used for generation
true_std = 15       # population standard deviation
data = np.random.normal(loc=true_mean, scale=true_std, size=sample_size)

# Hypothesized population mean to test against
hypothesized_mean = 105

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, hypothesized_mean)

# Output results
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"Hypothesized Mean: {hypothesized_mean}")
print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the hypothesized mean.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference from the hypothesized mean.")

# Visualize the distribution
plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, edgecolor='black', alpha=0.7)
plt.axvline(np.mean(data), color='blue', linestyle='--', label=f'Sample Mean = {np.mean(data):.2f}')
plt.axvline(hypothesized_mean, color='red', linestyle='-', label=f'Hypothesized Mean = {hypothesized_mean}')
plt.title('Histogram of Simulated Data with Mean Comparison')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2

# Sample data (e.g., quality control measurements)
data = [9.2, 8.7, 9.4, 8.9, 9.1, 8.8, 9.3, 9.0, 8.6, 9.1]

# Parameters
n = len(data)                             # sample size
sample_variance = np.var(data, ddof=1)   # sample variance
hypothesized_variance = 0.04             # claimed population variance (e.g., σ² = 0.2²)

# Chi-square statistic
chi_square_stat = (n - 1) * sample_variance / hypothesized_variance

# Degrees of freedom
df = n - 1

# Significance level
alpha = 0.05

# Critical values for two-tailed test
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)

# Output
print(f"Sample Size (n): {n}")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Hypothesized Variance: {hypothesized_variance}")
print(f"Chi-Square Statistic: {chi_square_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Chi-Square Critical Values (α = {alpha}): Lower = {chi2_lower:.4f}, Upper = {chi2_upper:.4f}")

# Interpretation
if chi_square_stat < chi2_lower or chi_square_stat > chi2_upper:
    print("Result: Reject the null hypothesis. Population variance is significantly different from the hypothesized value.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference in population variance.")
```

```python
#Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Example data:
# Group A: 45 successes out of 100 trials
# Group B: 30 successes out of 100 trials

successes = np.array([45, 30])   # Number of successes in each group
samples = np.array([100, 100])   # Number of observations (sample size) for each group

# Perform two-proportion Z-test
z_stat, p_value = proportions_ztest(successes, samples)

# Output
print(f"Group A: {successes[0]} successes out of {samples[0]}")
print(f"Group B: {successes[1]} successes out of {samples[1]}")
print(f"\nZ-Statistic: {z_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. Proportions are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference in proportions.")
```

```python
#Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Sample data (e.g., test scores from two different schools)
data1 = [85, 88, 90, 86, 87, 89, 84, 88, 85, 87]
data2 = [75, 78, 80, 76, 77, 82, 79, 74, 73, 76]

# Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# F-statistic (larger variance / smaller variance)
if var1 > var2:
    f_stat = var1 / var2
    dfn, dfd = len(data1) - 1, len(data2) - 1
else:
    f_stat = var2 / var1
    dfn, dfd = len(data2) - 1, len(data1) - 1

# Significance level
alpha = 0.05

# Critical F values (two-tailed test)
f_critical_low = f.ppf(alpha / 2, dfn, dfd)
f_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)

# P-value for right-tailed test (use two-tailed logic)
p_value = 2 * min(1 - f.cdf(f_stat, dfn, dfd), f.cdf(f_stat, dfn, dfd))

# Output results
print(f"Sample Variance 1: {var1:.4f}")
print(f"Sample Variance 2: {var2:.4f}")
print(f"F-Statistic: {f_stat:.4f}")
print(f"Degrees of Freedom: ({dfn}, {dfd})")
print(f"P-Value: {p_value:.4f}")
print(f"Critical F-Values (α = {alpha}): Lower = {f_critical_low:.4f}, Upper = {f_critical_high:.4f}")

# Interpretation
if f_stat < f_critical_low or f_stat > f_critical_high:
    print("Result: Reject the null hypothesis. Variances are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference in variances.")

# Visualization of F-distribution
x = np.linspace(0.1, 5, 1000)
y = f.pdf(x, dfn, dfd)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'F-distribution (df={dfn},{dfd})', color='blue')
plt.axvline(f_stat, color='orange', linestyle='--', label=f'F-statistic = {f_stat:.2f}')
plt.axvline(f_critical_low, color='red', linestyle='--', label='Lower Critical Value')
plt.axvline(f_critical_high, color='red', linestyle='--', label='Upper Critical Value')
plt.title('F-Test: Comparison of Two Variances')
plt.xlabel('F Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulate observed data (e.g., outcomes of a die rolled 120 times)
np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=120)
observed_counts = np.array([np.sum(observed == i) for i in range(1, 7)])

# Expected frequencies assuming fair die (uniform distribution)
expected_counts = np.array([120 / 6] * 6)  # 20 for each face

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Output
print("Observed Frequencies:", observed_counts)
print("Expected Frequencies:", expected_counts)
print(f"\nChi-Square Statistic: {chi2_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The observed data does not fit the expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis. The observed data fits the expected distribution.")

# Visualization: Observed vs Expected frequencies
faces = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(faces))

plt.figure(figsize=(8, 5))
plt.bar(x - 0.15, observed_counts, width=0.3, label='Observed', color='skyblue')
plt.bar(x + 0.15, expected_counts, width=0.3, label='Expected', color='orange')
plt.xticks(x, faces)
plt.title('Observed vs Expected Frequencies (Die Rolls)')
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
